```julia
using JLD2
using Random
import Tensor_FixedSeedMC as qo
using Statistics
using Plots
using LinearAlgebra
```

## 圧縮しない場合(TCI)

```julia
JLD2.@load "Ks_sigmas_stock0_10_5_american_bond10.JLD2" tt_op
tt_tci = tt_op.sitetensors
```

```julia
#option_tt = itensor_toArray(2, M_tci)
option_tt = tt_tci
@show size(option_tt[1])
@show size(option_tt[2])
@show size(option_tt[3])
```

```julia
σs = collect(0.15:0.001:0.249)
len_σs = length(σs)
d = 3

num_samples = 100
random_combinations = qo.generate_random_combinations(len_σs, d, num_samples)

result_hako = []
result_time = []
for i in random_combinations
    #@show i
    time_inner = @elapsed begin
        inner_ = qo._evaluate(option_tt, i)
        result = real(inner_)
    end
    push!(result_hako, result)
    push!(result_time, time_inner)
end

```
